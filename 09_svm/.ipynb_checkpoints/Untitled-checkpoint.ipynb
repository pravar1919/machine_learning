```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.target
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df["target"] = iris.target
df
```

```python
df_0 = df[df.target==0]
df_1 = df[df.target==1]
df_2 = df[df.target==2]
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.scatter(df_0['sepal length (cm)'], df_0['sepal width (cm)'], color='red', marker='+')
plt.scatter(df_1['sepal length (cm)'], df_1['sepal width (cm)'], color='blue', marker='.')
# plt.scatter(df_2['sepal length (cm)'], df_2['sepal width (cm)'], color='green', marker='*')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis='columns'), df['target'], test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.svm import SVC
model = SVC(C=40)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
